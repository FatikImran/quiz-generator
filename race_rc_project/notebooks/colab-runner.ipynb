## 3. Inference Test
lets trace the models making a prediction.

```
import pandas as pd
from src.inference import predict_from_race_row

# load a sample row
print("loading val data..")
val_df = pd.read_csv("data/raw/val.csv")
sample_row = val_df.iloc[0].to_dict()

print("\nSample Passage:")
print(str(sample_row['article'])[:300] + "...")

result = predict_from_race_row(sample_row)

print("\n\n== Inference Output ==")
print(f"Generated Question: {result['generated_question']['question']}")
print(f"Predicted Answer: {result['verification']['predicted']}")
print(f"Actual Answer: {result['correct_label']}")

print("\nDistractors generated:")
for i, d in enumerate(result['distractors']):
    print(f" {i+1}. {d}")

print("\nGraduated Hints generated:")
for i, h in enumerate(result['hints']):
    print(f" Hint {i+1}: {h}")
```
